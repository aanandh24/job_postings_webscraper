```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re
import time
from urllib.error import HTTPError

def crawl_page(url):
    try:
        # Send a request to the URL
        req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
        response = urlopen(req)

        # Read the HTML content of the page
        html_content = response.read()

        # Create a BeautifulSoup object to parse the HTML
        soup = BeautifulSoup(html_content, 'html.parser')

        # Extract the information you need from the page
        # For example, print all the links on the page
        for link in soup.find_all('a'):
            print(link.get('href'))

    except HTTPError as e:
        print(f"Error accessing the URL: {url}")
        print(f"Error message: {e}")

    # Pause for a few seconds before crawling the next page
    time.sleep(2)
```

```python
url = "https://example.com"
crawl_page(url)
```

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re
import time
from urllib.error import HTTPError
import requests

try:
    from googlesearch import search
except ImportError:
    print("No module named 'google' found")
 
# to search
query = "site:boards.greenhouse.io data science intern"
query += " -DUBAI"

count = 0
results = search(query, tld="co.in", num=100, stop=100, pause=2) 
for j in results:
    time.sleep(1.0)
    url = j
    #print(url)
    dashes = url.count('/')
    if dashes <= 3:
        continue
        #print(url)
    position, company, location = '', '', ''

    
    try:
        response = requests.get(url)
    except Exception:
        continue
    # Create a BeautifulSoup object to parse the HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    element = soup.find('h1', class_='app-title')
    if element is not None:
        position = element.text.strip()
    
    # Find the desired element using its class attribute
    element = soup.find('span', class_='company-name')
    if element is not None:
        company = element.text.strip()

    # Find the desired element using its class attribute
    element = soup.find('div', class_='location')
    if element is not None:
        location = element.text.strip()
        
    desc = "{0} - {1} - {2}".format(company, position, location).upper()
    terms_to_exclude = list()
    terms_to_exclude += ['TORONTO', 'SINGAPORE', 'TUCOWS']
    
    skip = 0
    for term in terms_to_exclude:
        if term in desc:
            skip = 1
            break
    
    if skip:
        continue
    
    print('----------------------------------------')
    print('url:', url)
    print('\tcompany:', company)
    print('\tposition:', position)
    print('\tlocation:', location)
    print('----------------------------------------')
    print('')
    
    
    count += 1
    
    
```
